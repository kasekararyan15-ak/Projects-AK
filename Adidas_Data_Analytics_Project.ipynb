## Section 1: Read and Load the Dataset


```python
# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set visualization styles
plt.style.use('default')
sns.set_palette("husl")
sns.set_style("whitegrid")

print("✓ All libraries imported successfully!")
```

## Section 2: Data Cleaning and Preprocessing


```python
# Step 1: Check for Missing Values
print(" CHECKING FOR MISSING VALUES:")
missing_values = df.isnull().sum()
print(missing_values)
print(f"\n✓ Total missing values: {df.isnull().sum().sum()}")
print("\n" + "="*80)
```

```python
# Step 2: Check for Duplicate Rows
print(" CHECKING FOR DUPLICATE ROWS:")
duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {duplicates}")

if duplicates > 0:
    print(f"Removing {duplicates} duplicate rows...")
    df = df.drop_duplicates()
    print(f"✓ Duplicates removed! New shape: {df.shape}")
else:
    print("✓ No duplicates found!")
print("\n" + "="*80)
```

```python
# Step 3: Clean Column Names and Data
print(" CLEANING COLUMN NAMES:")
print(f"Original columns: {df.columns.tolist()}")

# Standardize column names (lowercase, remove extra spaces)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(f"Cleaned columns: {df.columns.tolist()}")
print("\n" + "="*80)
```

```python
# Step 4: Check Data Types
print(" CURRENT DATA TYPES:")
print(df.dtypes)
print("\n" + "="*80)
```

```python
# Step 5: Handle Missing Values (if any)
print(" HANDLING MISSING VALUES:")
# For numeric columns: fill with median
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)
        print(f"  Filled {col} with median value")

# For categorical columns: fill with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)
        print(f"  Filled {col} with mode value")

print("✓ Missing values handled!")
print("\n" + "="*80)
```

```python
# Step 6: Detect and Handle Outliers using IQR method
print(" DETECTING OUTLIERS:")
outlier_count = 0

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    if col_outliers > 0:
        print(f"  {col}: {col_outliers} outliers found")
        outlier_count += col_outliers

print(f"\n✓ Total outliers detected: {outlier_count}")
print("  Note: Outliers are kept for analysis (represent legitimate extreme values)")
print("\n" + "="*80)
```

```python
# Step 7: Display Cleaned Data Summary
print(" DATA CLEANING COMPLETE!")
print(f"\n Final Dataset Shape: {df.shape[0]} rows × {df.shape[1]} columns")
print("\n Final Missing Values Count:")
print(df.isnull().sum())
print("\n" + "="*80)
```

## Section 3: Exploratory Data Analysis (EDA)


```python
# Step 1: Get numeric and categorical columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

print(" NUMERIC COLUMNS:")
print(numeric_columns)
print("\n CATEGORICAL COLUMNS:")
print(categorical_columns)
print("\n" + "="*80)
```

```python
# Step 2: Distribution Analysis - Numeric Columns (Histograms)
print(" DISTRIBUTION ANALYSIS - NUMERIC VARIABLES")
if numeric_columns:
    fig, axes = plt.subplots(nrows=(len(numeric_columns)+1)//2, ncols=2, figsize=(14, 4*(len(numeric_columns)+1)//2))
    axes = axes.flatten() if len(numeric_columns) > 1 else [axes]
    
    for idx, col in enumerate(numeric_columns):
        axes[idx].hist(df[col], bins=30, color='steelblue', edgecolor='black', alpha=0.7)
        axes[idx].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('Frequency')
        axes[idx].grid(alpha=0.3)
    
    # Hide unused subplots
    for idx in range(len(numeric_columns), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()
else:
    print("No numeric columns to visualize")
print("\n" + "="*80)
```

```python
# Step 3: Box Plots for Numeric Columns
print(" BOX PLOT ANALYSIS - NUMERIC VARIABLES")
if numeric_columns:
    fig, axes = plt.subplots(nrows=(len(numeric_columns)+1)//2, ncols=2, figsize=(14, 4*(len(numeric_columns)+1)//2))
    axes = axes.flatten() if len(numeric_columns) > 1 else [axes]
    
    for idx, col in enumerate(numeric_columns):
        axes[idx].boxplot(df[col], vert=True)
        axes[idx].set_title(f'Box Plot of {col}', fontsize=12, fontweight='bold')
        axes[idx].set_ylabel(col)
        axes[idx].grid(alpha=0.3)
    
    # Hide unused subplots
    for idx in range(len(numeric_columns), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()
else:
    print("No numeric columns to visualize")
print("\n" + "="*80)
```

```python
# Step 4: Categorical Analysis - Value Counts
print(" CATEGORICAL VARIABLES ANALYSIS - VALUE COUNTS")
for col in categorical_columns:
    print(f"\n{col.upper()}:")
    print(df[col].value_counts())
    print("-" * 40)
print("\n" + "="*80)
```

```python
# Step 5: Bar Charts for Categorical Variables
print(" BAR CHARTS - CATEGORICAL VARIABLES")
if categorical_columns:
    fig, axes = plt.subplots(nrows=(len(categorical_columns)+1)//2, ncols=2, figsize=(14, 4*(len(categorical_columns)+1)//2))
    axes = axes.flatten() if len(categorical_columns) > 1 else [axes]
    
    for idx, col in enumerate(categorical_columns):
        value_counts = df[col].value_counts()
        axes[idx].bar(range(len(value_counts)), value_counts.values, color='teal', alpha=0.7, edgecolor='black')
        axes[idx].set_xticks(range(len(value_counts)))
        axes[idx].set_xticklabels(value_counts.index, rotation=45, ha='right')
        axes[idx].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
        axes[idx].set_ylabel('Frequency')
        axes[idx].grid(alpha=0.3, axis='y')
    
    # Hide unused subplots
    for idx in range(len(categorical_columns), len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()
else:
    print("No categorical columns to visualize")
print("\n" + "="*80)
```

```python
# Step 6: Correlation Analysis
print(" CORRELATION HEATMAP - NUMERIC VARIABLES")
if numeric_columns:
    plt.figure(figsize=(12, 8))
    correlation_matrix = df[numeric_columns].corr()
    sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
                square=True, linewidths=1, cbar_kws={"shrink": 0.8})
    plt.title('Correlation Matrix of Numeric Variables', fontsize=14, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.show()
    
    print("\n KEY CORRELATIONS:")
    for i in range(len(correlation_matrix.columns)):
        for j in range(i+1, len(correlation_matrix.columns)):
            corr_value = correlation_matrix.iloc[i, j]
            if abs(corr_value) > 0.5:  # Show strong correlations
                print(f"  {correlation_matrix.columns[i]} <-> {correlation_matrix.columns[j]}: {corr_value:.3f}")
else:
    print("No numeric columns for correlation analysis")
print("\n" + "="*80)
```

```python
# Step 7: Scatter Plots for Key Relationships
print(" SCATTER PLOTS - BIVARIATE ANALYSIS")
if len(numeric_columns) >= 2:
    # Create scatter plots for first few numeric column pairs
    fig, axes = plt.subplots(nrows=min(2, (len(numeric_columns)-1)//2 + 1), 
                             ncols=2, figsize=(14, 8))
    axes = axes.flatten() if len(numeric_columns) > 1 else [axes]
    
    plot_idx = 0
    for i in range(len(numeric_columns)-1):
        for j in range(i+1, len(numeric_columns)):
            if plot_idx < len(axes):
                axes[plot_idx].scatter(df[numeric_columns[i]], df[numeric_columns[j]], 
                                      alpha=0.6, color='steelblue', edgecolor='black', s=50)
                axes[plot_idx].set_xlabel(numeric_columns[i])
                axes[plot_idx].set_ylabel(numeric_columns[j])
                axes[plot_idx].set_title(f'{numeric_columns[i]} vs {numeric_columns[j]}', 
                                        fontsize=11, fontweight='bold')
                axes[plot_idx].grid(alpha=0.3)
                plot_idx += 1
    
    # Hide unused subplots
    for idx in range(plot_idx, len(axes)):
        axes[idx].set_visible(False)
    
    plt.tight_layout()
    plt.show()
else:
    print("Need at least 2 numeric columns for scatter plots")
print("\n" + "="*80)
```

```python
# Step 8: Statistical Summary and Insights
print(" STATISTICAL INSIGHTS FROM EDA")
print("\n NUMERIC COLUMNS SUMMARY:")
print(df[numeric_columns].describe().round(2))

print("\n\n KEY FINDINGS FROM DATA:")
print(f"  • Dataset contains {df.shape[0]} records with {df.shape[1]} features")
print(f"  • Numeric columns: {len(numeric_columns)}")
print(f"  • Categorical columns: {len(categorical_columns)}")

# Find columns with highest variance
if numeric_columns:
    variance_cols = df[numeric_columns].var().sort_values(ascending=False)
    print(f"\n  • Column with highest variance: {variance_cols.index[0]} ({variance_cols.values[0]:.2f})")
    print(f"  • Column with lowest variance: {variance_cols.index[-1]} ({variance_cols.values[-1]:.2f})")

print("\n" + "="*80)
```
